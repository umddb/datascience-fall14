```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sleep.csv')
```

```python
sleep = data.loc[:,['extra', 'group']]
group1 = sleep[data.group == 1]
group2 = sleep[data.group == 2]
```

```python
plt.title('Extra sleep experiment data')
plt.xlabel('Subject')
plt.ylabel('Extra sleep gained')
plt.plot(group1.extra, label = 'Drug 1')
plt.plot(group2.extra, label = 'Drug 2')
plt.legend(loc = 'upper right')
```
